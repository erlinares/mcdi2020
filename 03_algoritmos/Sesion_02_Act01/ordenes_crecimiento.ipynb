## Introducción
+ Un algoritmo es una serie de pasos finitos para resolver un problema.
+ La eficiencia de un algoritmo mide el consumo de recursos para resolver el problema.
+ La notación asimptótica nos sirve para representar la eficiencia de un algoritmo, para ver como aumenta el tiempo de ejecución de un algoritmo con el tamaño de la entrada.
+ Un algoritmo que sea asintóticamente más eficiente será la mejor opción para todas las entradas excepto para muy pequeñas.

Los ordenes de crecimiento nos indican como se incrementa el tiempo de ejecución conforme el tamaño de la entrada crece.
+ Si un algoritmo es constante, su notación  O "grande" es O(1), lo que indica que no depende del tamaño de la entrada.
+ Si un algoritmo es lineal su tiempo de ejecución es n, se representa como  O(n), lo que indica que crece a la misma velocidad.
+ De esta manera hay difrentes funciones que representan las ordenes de crecimiento de un algoritmo.

En las siguientes gráficas comparativas se mostrará el orden de crecimiento de dos diferentes funciones y se hará una interpretación de las mismas  



## Comparativa de las Ordenes de Crecimiento 

## 1. Funciones $O(1)$   vs   $O( log n) $ 

```python
#Graphics O(1) vs O(log n)
import matplotlib.pyplot as plt 
import math 
from datetime import datetime

t = range(2,100000,2)
y1 = []
y2 = []
for n in t:
    y1.append(math.log(n,2))
    y2.append(n/n)
%matplotlib inline 
plt.figure(figsize=(15,5))    
p1, p2 = plt.plot(t, y1, t, y2) 
texto1 = plt.text(90000, 2, r'O(1)', fontsize=20)
texto2 = plt.text(90000, 14, r'O(log n)', fontsize=16)
plt.grid()
plt.title('Representacion de las funciones: O(1) vs O(log n)')
plt.ylabel('Operaciones') 
plt.xlabel('Entradas') 
plt.show() 
```

# Análisis $ O(1)$   vs   $O( log n) $
+ $O(1)$ nos muestra un crecimiento constante, es decir que no depende del tamaño de la entrada.
+ $O( log n) $, vemos una velocidad de crecimiento más rápida comparada con $O(1)$, no obstante esto, llega un momento en que empieza a ser constante conforme se incrementa el número de entrada
+ Si tenemos un algoritmo con un orden de crecimiento $ O(1)$ sería lo ideal para su implementación.

## 2. Funciones $ O(n)$   vs   $O( n log n) $

```python
t = range(2, 100000, 2)
y1 = []
y2=[]
for n in t:
    y1.append(n*math.log(n))
    y2.append(n)
%matplotlib inline 
plt.figure(figsize=(15,5))    
p1, p2 = plt.plot(t, y1, t, y2)
texto1 = plt.text(90000, 100000, r'O(n)', fontsize=20)
texto2 = plt.text(90000, 900000, r'O(n log n)', fontsize=16)
plt.grid()
plt.title('Representacion de las funciones: O(n)   vs   O( n log n)')
plt.ylabel('Operaciones') 
plt.xlabel('Entradas') 
plt.show()
```

## Análisis $ O(n)$   vs   $O( n log n) $
+ $ O(n)$,  muestra un crecimiento lineal, es decir que crece conforme el tamaño de la entrada aumenta.
+ $O(n log n) $, muestra un crecimiento lineal pero más veloz de $O(n)$.
+ El algoritmo con un orden de crecimiento $ O(n)$ es lo ideal para su implementación.

## 3. Funciones $ O(n^{2})$   vs  $O(n^{3})$

```python
#Graphics O(n**2) vs O(n**3)
t = range(1,10,1)
y1 = []
y2=[]
for n in t:
    y1.append(n**2)
    y2.append(n**3)
%matplotlib inline
plt.figure(figsize=(15,5))    
p1, p2 = plt.plot(t, y1, t, y2)
texto1 = plt.text(9, 100, r'$O(n^{2})$', fontsize=20)
texto2 = plt.text(9, 700, r'$O(n^{3})$', fontsize=16)
plt.grid()
plt.title('Representacion de las funciones: $O(n^{2})$   vs   $O(n^{3})$')
plt.ylabel('Operaciones') 
plt.xlabel('Entradas') 
plt.show()
```

## Análisis $ O(n^{2})$   vs  $O(n^{3})$
+ $ O(n^{2})$,  muestra un crecimiento cuadrado, es decir que crece al cuadrado conforme el tamaño de la entrada aumenta.
+ $ O(n^{3})$, muestra un crecimiento cúbico, y más veloz que  $O(n^{2})$.
+ El algoritmo con un orden de crecimiento $ O(n^{3})$ no es el ideal para su implementación en este caso

## 4. Funciones $ O(a^{n})$   vs  $O(n!)$

```python
#Graphics a**n vs n!
t = range(1,7,1)
y1 = []
y2=[]
for n in t:
    #corregir formula 
    y1.append(2.7**n)
    y2.append(math.factorial(n))
%matplotlib inline
plt.figure(figsize=(10,5))    
p1, p2 = plt.plot(t, y1, t, y2)
texto1 = plt.text(6, 100 , r'$a^{n}$', fontsize=20)
texto2 = plt.text(6, 700, r'n!', fontsize=16)
plt.grid()
plt.title('Representacion de las funciones: $ O(a^{n})$   vs  $O(n!)$')
plt.ylabel('Operaciones') 
plt.xlabel('Entradas') 
plt.show() 
```

## Análisis $ O(a^{n})$   vs  $O(n!)$
+ $ O(a^{n})$ y  $O(n!)$,  muestra un crecimiento exponencial muy acelerado, es decir que crecen mucho conforme el tamaño de la entrada aumenta.
+ Sin embargo el algoritmo con un orden de crecimiento $ O(a^{n})$, es el ideal en este caso para su implementación.

## 5. Funciones $n!$   vs  $n^{n}$


```python
#Graphics n! vs n**n
t = range(1,7,1)
y1 = []
y2=[]
for n in t:
    y1.append(math.factorial(n))
    y2.append(n**n)
%matplotlib inline
plt.figure(figsize=(10,5))    
p1, p2 = plt.plot(t, y1, t, y2)
texto1 = plt.text(6, 100 , r'n!', fontsize=20)
texto2 = plt.text(6, 40000, r'$n^{n}$', fontsize=16)
plt.grid()
plt.title('Representacion de las funciones: $ n!$   vs  $n^{n}$')
plt.ylabel('Operaciones') 
plt.xlabel('Entradas') 
plt.show() 
```

## Análisis $n!$   vs  $n^{n}$
+ $n^{n}$ es la función con un orden de crecimiento demasiado veloz
+ Aunque ambas funciones con de una velocidad de crecimiento grande, un aloritmo con una función de crecimiento $n^{n}$ sería el menos adecuado para su implementación, ya que conforme aumentada su entrada aumentará el tiempo de ejecución.

# Conclusión
+ Las gráficas de representación del orden de crecimiento, nos permitieron observar la eficacia de cada función con respecto a un algoritmo.
+ Lo ideal es tener algoritmos con un crecimiento constante que no dependa del tamaño de la entrada, como $O(1)$.
+ Funciones de crecimiento como $O(1)$ , $ O(n)$ y $O( log n) $ son las mejores porque su orden de crecimiento es mas lento.
+ Funciones de crecimiento como $ O(n^{2})$, $O(n!)$ y $n^{n}$ son las menos deseables ya que su orden de crecimiento es mas rápido.

## Simulaciones
+ La simulación de las funciones nos permite ver que entre mayor sea la entrada, las funciones de crecimiento más veloces como $O(n^{2})$, $O(n!)$ y $n^{n}$ se elevan excesivamente con respecto a las ordenes de crecimiento más lentos como $O(1)$ , $ O(n)$ y $O( log n) $, lo que nos indica la importancia de evaluar la eficiencia de un algitmo antes de su implementación ya que con entradas pequeñas funcionarán relativamente bien, pero cuando las entradas incrementan su orden de crecimiento es más acelerado.
+ La Simulación 9.4 de las Funciones $ O(a^{n})$   vs  $O(n!)$  n=100000, no se ejecuto debido a que python al generar los factoriales se tardaba mucho, por lo que seguramente será necesario usar Julia para hacer pruebas.
+ De igual manera, las simulaciones de las funciones  $n!$   vs  $n^{n}$  n=1000 y n=10000, solo se ejecuto y guardo en un arreglo porque como se muestra los valores factoriales y de $n^{n}$ son excesivamente grandes para poderlos graficar.
+ Y la simulación  de $n^{n}$  n=1000 y n=100000 no se ejecuto por el motivo del tiempo de ejecución para el calculo de los factoriales y las potencias.
