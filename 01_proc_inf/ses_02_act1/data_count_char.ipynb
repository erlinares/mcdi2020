## Código

```python
# import library
import collections
import operator
import matplotlib.pyplot as plt 

#create array with files names
arrTxtFiles = ["text_1.txt","text_2.txt","text_3.txt","text_4.txt", "text_5.txt"]
```

```python
#iterate array, count characters from each file
for i in arrTxtFiles: 
    #open file and count characters
    textFile = open(i, "r", encoding="UTF-8")
    countTxt = collections.Counter(textFile.read())
    #sort collections.Counter, put in  list
    sortList=sorted(countTxt.items())
    #create dictionary from list
    sortDict=dict(sortList)
    print ( countTxt.most_common(20) )
    # create list x, y
    x = list (sortDict.keys())
    y = list (sortDict.values())
    #draw frequency
    %matplotlib inline 
    plt.figure(figsize=(15,3))    
    plt.bar(x,y) 
    plt.title('Frecuencia de caracteres - ' + i) 
    plt.ylabel('Frecuencia') 
    plt.xlabel('Caracter') 
    plt.show()
```

## Interpretación
Analizando las gráficas se puede observar lo siguiente:  
+ En "Frecuencia de caracteres - text1.txt" el caracter que más se repite es la letra **a**.
+ En "Frecuencia de caracteres - text2.txt" el caracter que más se repite es la letra **e**.
+ En "Frecuencia de caracteres - text3.txt" el caracter que más se repite es la letra **i**.
+ En "Frecuencia de caracteres - text4.txt" el caracter que más se repite es la letra **o**.
+ En "Frecuencia de caracteres - text5.txt" el caracter que más se repite es la letra **u**.
+ En todas las gráficas el **espacio** es el segundo caracter que más se repite.
+ En todas las gráficas se cuentan los signos de puntuación y caracteres especiales a (.,()!¡¿?-.)
+ En todas las gráficas no hay distinción de mayúsculas y mínusculas ejemplo: (**B**,**b**) 
+ En todas las gráficas no hay distinción de acentos ejemplo: (**a**,**á**)
+ Las consonantes más utilizadas en los textos son (r,s,t,l,c,n)  

De lo anterior se concluye que estamos observando textos monovocálicos (donde se emplea sólo una vocal), y  para  un mejor análisis es necesario eliminar caracteres especiales y los signos de puntuación y números, así como unficiar  caracteres (mayúsculas, minúsculas y acentuadas).
